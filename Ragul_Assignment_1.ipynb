## 7. Concatenate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
res=np.concatenate((a,b),axis=0)
print(res)
```
